## 1. What is supervised learning ?

** Let's define machine learning first **

Machine learning intends to give computers the ability to learn without being explicitely programmed by a manually defined set of rules. <br/>

While computers can follow strict programming instructions, machine learning aims to develop algorithms that learn from data. <br/>

Algorithms are constructed in order to learn from relationships and trends in the data. <br/>

** What is Supervised Learning ?**

Supervised learning is a field in machine learning that intends to learn from labeled training data. <br/>

Labeled training data is a set of historical observations. Each observation is composed of a vector of input values (features) and a vector of output values (labels). Supervised learning algorithms intends to analyse training data in order to infer a function that maps input values with output values. <br/> 

Inferred function can then be used to predict labels of new unseen observations. <br/>

**What is the difference between supervised learning and statistical inference ?**

Machine learning and statistical inference are indeed intertwined. <br/> 

They both rely on minimizing loss functions to learn parameters and developments in each field can influence the other. <br/>

While there are still discussions on drawing the exact frontiers, practicians tend to agree on the following differences:<br/>

1. Statistical inference is about drawing valid conclusions. Machine learning is about predictions.<br/>

2. Statistical inference relies heavily on data modeling and assumptions. Machine learning relies on the predictive accuracy of models.<br/>

3. Statistical inference emphazises low dimensional problems. Machine learning emphazises high dimensional problems.<br/>

4. Statistical inference takes advantage of asymptotics and parametrics models. Machine learning leverages computational power.<br/>


## 2. Methodology

** Steps **

Conducting a supervised learning task usually includes the following steps:

1. Identify a dataset of observations including both features and a label value.
2. Split observations in two datasets: training data and testing data. 
3. Train supervised learning algorithm on training data to infer a function
4. Use trained algorithm to predict labels of testing data
5. Evaluate predictive accuracy of algorithm by comparing predicted labels with observed labels of testing data

**Training accuracy and testing accuracy**

Predictive accuracy is measured by the mean accuracy between predictions and actual values. <br/>

Predictive accuracy on training data is called training accuracy.<br/>
Predictive accuracy on testing data is called testing accuracy.<br/>

While counter-intuitive, training accuracy can be useful to evaluate model ability to map features with labels in the training set. This is useful in the bias-variance tradeoff.

** Bias-variance tradeoff **

As supervised learning is about prediction it is often the case that increasing function complexity will lead to better performance on training error. <br/> However increasing complexity in order to perfectly fit shapes of the training set will prevent algorithms from generalizing beyond their training set. <br/>

This calls for the bias-variance tradeoff, which is the problem of minimizing the two sources of error when generalizing a function: <br/>

1. Bias: model misses relevant relationships between input features and target ouput
2. Variance: model becomes to sensitive to small variations specific to the training set and start modeling random noise in the training data.

High bias error is referred as 'underfitting'. <br/>
Hig variance error is referred as 'overfitting'.
