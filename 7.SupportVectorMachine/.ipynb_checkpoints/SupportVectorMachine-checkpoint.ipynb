## 1. What is a support vector machine ?

** What is support vector machine ? **

A support vector machine is a supervised machine learning algorithm that can be used both for classification and regression problems. <br/>

It is however mainly used in classification problems. <br/>

** How does support vector machine work ? **

Having labeled data in a n-dimensional space, SVM will find the hyper-plane that best differentiates the classes.

### 2.2 Soft-margin

Soft-margin SVM allows for some observations to be wrongly classified. <br\>

Without entering into mathematical details, soft-margin SVM introduces a trade-off between uncorrectly classified observations and model stability. <br\>

Parameter C in the soft-margin cost function controls the cost of misclassiying an observation in the definition of the classifier. 

- Low C values mean low cost for misclassified observations <br\>
- High C values mean high cost for misclassified observations

## 3. Non linear SVM

Let's now consider the following distribution for our gender classification problem.

```python
from IPython.display import Image
Image(filename='SVM4.png',width=300, height=300)
```

**3.1 Can you draw hyper-planes that split men and women, based on hair length and height features ?**

Practically almost impossible to find such a linear hyper-plane

In practice, most of classification problems are non linear. <br\>

Non linear SVM allows to draw non linear hyper-planes fitting non linear classifications. <br\>

Without entering mathematical considerations, non linear SVM replaces dot product by non linear kernel functions (Gaussian radial basis function). <br\>

In non linear SVM, gamma is the free parameter of the Gaussian radial basis function. 

- Low gamma implies observations'class will have influence on other observations even if distance between them is large. <br\>

- High gamma implies observations'class only influence observations within short distance. <br\>

**Yeaaaah... C and Gamma sounds cool, but I am not sure I get it ...**

C parameter tells the SVM optimization the cost of misclassifying an observation. <br\>
Low values mean ‘low cost’ and high values mean ‘high cost’. <br\>

Gamma parameter defines how far the influence of a single training example reaches. <br\>
Low values mean ‘far’ and high values mean ‘close’. <br>

Increasing C increases the cost of misclassification: <br>
- ... variance
- ... bias

Increasing gamma reduce reach of influence of support vectors: <br>
- ... variance
- ... bias

Visualise the SVM classification by running Visualisation code below.

**4.3 Comment the results ? **

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))

plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC classifier')
plt.show()
```

**4.11 Can you get the optimal configuration for iris species prediction ? **

Training and testing data

```python
X = iris.data[:, :4]
y = iris.target

train_features = []
train_targets = []
test_features = []
test_targets = []

for i in xrange(len(X)):
    if i % 2 == 0:
        train_features.append(X[i])
        train_targets.append(y[i])
    if i % 2 != 0:
        test_features.append(X[i])
        test_targets.append(y[i])
```

```python
from sklearn import svm

train_accuracies = []
test_accuracies = []
configuration = []

gamma = [0,1,2,5,10,20,30]
C = [1]

for c_value in C:
    for gamma_value in gamma:    
        svc = svm.SVC(kernel='rbf', C=c_value,gamma=gamma_value).fit(train_features, train_targets)
        train_accuracies.append(svc.score(train_features,train_targets))
        test_accuracies.append(svc.score(test_features,test_targets))
        configuration.append("gamma:%s,C:%s"%(gamma_value, c_value))
        
print configuration
```

Run the following lines to visualize the learning plot

```python
import matplotlib.pyplot as plt
plt.axis([0,6, 0.5, 1.1])
plt.plot(train_accuracies, 'k')
plt.plot(test_accuracies, 'g')
plt.title('accuracies')

plt.show()
```
